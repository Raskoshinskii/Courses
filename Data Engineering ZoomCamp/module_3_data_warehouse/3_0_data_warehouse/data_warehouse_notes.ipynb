### Partitioning and Clustering

**Partitioning Advantages**
- Improved Peformance: limitting amount of data that need to be scanned or processed. 
- Parallelism: Partitioning enables parallel processing of data across multiple partitions. By distributing workload across multiple CPU cores or nodes in a distributed system, partitioning can significantly improve throughput and reduce query latency.
- Scalability: Partitioning supports horizontal scalability by allowing data to be distributed across multiple physical or logical storage units.
- Better Data Management: data is grouped into logical units (days, customer segments, geo ...)

**Clustering Advantages**
- Improves queries filtering (faster)
 
**Important**
- Tables that are less than 1GB don't show significant improvement by using partitioning and clustering =
- There is usually a limitation on number of partitions on clusters

**Clustering over Partitioning**
- When we have small data per partition
- When it results in a large number of partitions 
- When modification operations happen too often with partitions (every few minutes)
